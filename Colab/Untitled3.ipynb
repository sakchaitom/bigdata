```python
dictionary_example = {
  "name": "name1 surname1",
  "age": 25,
  "gender": "male"
}

print(dictionary_example["name"])
print(dictionary_example["age"])
print(dictionary_example["gender"])
```

**List Dictionary**

```python
dict_in_list = [{
  "name": "name1 surname1",
  "age": 25,
  "gender": "male",
  "tel": ["11111","22222",{
      "home": "3333",
      "office": "4444"
  }]
},{
  "name": "name2 surname2",
  "age": 50,
  "gender": "female"
}]


print(dict_in_list[0]["name"])
print(dict_in_list[0]["age"])
print(dict_in_list[0]["gender"])

print(dict_in_list[0]["tel"][2]["home"])
```
