```
import pandas as pd
import numpy as np
import joblib
import pickle
from attacut import tokenize
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from google.colab import files
```

```
url = 'https://nriis.go.th/temp/strategy.csv'
df = pd.read_csv(url)
```

```
df.count()
```

```
df.info()
```

```
y_df = df['feature']
```

```
y_df
```

```
x_df = df['name']
```

```
x_df
```

```
count_vect = CountVectorizer(tokenizer=tokenize)
```

```
Xtrain_count = count_vect.fit_transform(x_df)
```

```
Xtrain_count.shape
```

```
Xtrain_count[0].toarray()
```

```
tf_transformer = TfidfTransformer(use_idf=False)
```

```
Xtrain_tf = tf_transformer.transform(Xtrain_count)
```

```
Xtrain_tf.shape
```

```
Xtrain_tf[0].toarray()
```

```
X_train, X_test, y_train, y_test = train_test_split(Xtrain_count, y_df, test_size=0.33, random_state=42)
```

```
clf = MultinomialNB()
```

```
clf.fit(X_train, y_train)
```

##ประเมินประสิทธิภาพของ **Model**

```
clf.score(X_test,y_test)
```

```
y_pred = clf.predict(X_test)
```

```
print(classification_report(y_test, y_pred))
```
